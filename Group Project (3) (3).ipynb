```python
# Read data from external file 
import pandas as pd 
train3 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Train_fold_3.csv")
test3 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Test_fold_3.csv")
```

```python
# Data slicing 
xtrain3 = train3.iloc[:, 0:24] #features
ytrain3 = train3.iloc[:, 24:25] #label
xtest3 = test3.iloc[:, 0:24] #features
ytest3 = test3.iloc[:, 24:25] #label
```

```python
#Train-3
# Convert dataframe to array 
xtrain3 = xtrain3.values 
ytrain3 = ytrain3.values 
ytrain3 = ytrain3.ravel() 

# K-Nearest Neighbor
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
from sklearn.neighbors import KNeighborsClassifier 
ModelAKU = KNeighborsClassifier(n_neighbors=49, weights='distance', algorithm='auto', leaf_size=10, metric='hamming')

ModelAKU.fit(xtrain3, ytrain3) 
yPredtrain3 = ModelAKU.predict(xtrain3) 

# Testing & performance analysis -Training data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytrain3, yPredtrain3), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Training: Accuracy Score: {0}'.format(skm.accuracy_score(ytrain3, yPredtrain3)) 
plt.title(all_sample_title, size = 15);
```

```python
#Test-3
# Convert dataframe to array 
xtest3 = xtest3.values 
ytest3 = ytest3.values 
ytest3 = ytest3.ravel() 

yPredtest3 = ModelAKU.predict(xtest3) 

# Testing & performance analysis -Testing data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytest3, yPredtest3), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Testing: Accuracy Score: {0}'.format(skm.accuracy_score(ytest3, yPredtest3)) 
plt.title(all_sample_title, size = 15);
```
