```python
# Read data from external file 
import pandas as pd 
train1 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Train_fold_1.csv")
test1 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Test_fold_1.csv")
```

```python
# Data slicing 
xtrain1 = train1.iloc[:, 0:24] #features
ytrain1 = train1.iloc[:, 24:25] #label
xtest1 = test1.iloc[:, 0:24] #features
ytest1 = test1.iloc[:, 24:25] #label
```

```python
#Train-1
# Convert dataframe to array 
xtrain1 = xtrain1.values
ytrain1 = ytrain1.values 
ytrain1 = ytrain1.ravel() 

# K-Nearest Neighbor
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
from sklearn.neighbors import KNeighborsClassifier 
ModelAKU = KNeighborsClassifier(n_neighbors=49, weights='distance', algorithm='auto', leaf_size=10, metric='hamming') 

ModelAKU.fit(xtrain1, ytrain1) 
yPredtrain1 = ModelAKU.predict(xtrain1) 

# Testing & performance analysis -Training data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytrain1, yPredtrain1), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Training: Accuracy Score: {0}'.format(skm.accuracy_score(ytrain1, yPredtrain1)) 
plt.title(all_sample_title, size = 15);
```

```python
#Test-1
# Convert dataframe to array 
xtest1 = xtest1.values 
ytest1 = ytest1.values 
ytest1 = ytest1.ravel() 

yPredtest1 = ModelAKU.predict(xtest1) 

# Testing & performance analysis -Testing data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytest1, yPredtest1), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Testing: Accuracy Score: {0}'.format(skm.accuracy_score(ytest1, yPredtest1)) 
plt.title(all_sample_title, size = 15);
```
