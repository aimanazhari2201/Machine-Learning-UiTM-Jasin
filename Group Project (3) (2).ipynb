```python
# Read data from external file 
import pandas as pd 
train2 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Train_fold_2.csv")
test2 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Test_fold_2.csv")
```

```python
# Data slicing 
xtrain2 = train2.iloc[:, 0:24] #features
ytrain2 = train2.iloc[:, 24:25] #label
xtest2 = test2.iloc[:, 0:24] #features
ytest2 = test2.iloc[:, 24:25] #label
```

```python
#Train-2
# Convert dataframe to array 
xtrain2 = xtrain2.values 
ytrain2 = ytrain2.values 
ytrain2 = ytrain2.ravel() 

# K-Nearest Neighbor
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
from sklearn.neighbors import KNeighborsClassifier 
ModelAKU = KNeighborsClassifier(n_neighbors=8, weights='distance', algorithm='ball_tree', leaf_size=50, metric='hamming') 

ModelAKU.fit(xtrain2, ytrain2) 
yPredtrain2 = ModelAKU.predict(xtrain2) 

# Testing & performance analysis -Training data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytrain2, yPredtrain2), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Training: Accuracy Score: {0}'.format(skm.accuracy_score(ytrain2, yPredtrain2)) 
plt.title(all_sample_title, size = 15);
```

```python
#Test-2
# Convert dataframe to array 
xtest2 = xtest2.values 
ytest2 = ytest2.values 
ytest2 = ytest2.ravel() 

yPredtest2 = ModelAKU.predict(xtest2) 

# Testing & performance analysis -Testing data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytest2, yPredtest2), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Testing: Accuracy Score: {0}'.format(skm.accuracy_score(ytest2, yPredtest2)) 
plt.title(all_sample_title, size = 15);
```
