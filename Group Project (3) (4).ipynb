```python
# Read data from external file 
import pandas as pd 
train4 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Train_fold_4.csv")
test4 = pd.read_csv("C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Test_fold_4.csv")
```

```python
# Data slicing 
xtrain4 = train4.iloc[:, 0:24] #features
ytrain4 = train4.iloc[:, 24:25] #label
xtest4 = test4.iloc[:, 0:24] #features
ytest4 = test4.iloc[:, 24:25] #label
```

```python
#Train-4
# Convert dataframe to array 
xtrain4 = xtrain4.values 
ytrain4 = ytrain4.values 
ytrain4 = ytrain4.ravel() 

# K-Nearest Neighbor
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
from sklearn.neighbors import KNeighborsClassifier 
ModelAKU = KNeighborsClassifier(n_neighbors=4, weights='uniform', algorithm='auto', leaf_size=10, metric='hamming') 

ModelAKU.fit(xtrain4, ytrain4) 
yPredtrain4 = ModelAKU.predict(xtrain4) 

# Testing & performance analysis -Training data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytrain4, yPredtrain4), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Training: Accuracy Score: {0}'.format(skm.accuracy_score(ytrain4, yPredtrain4)) 
plt.title(all_sample_title, size = 15);
```

```python
#Test-4
# Convert dataframe to array 
xtest4 = xtest4.values 
ytest4 = ytest4.values 
ytest4 = ytest4.ravel() 

yPredtest4 = ModelAKU.predict(xtest4) 

# Testing & performance analysis -Testing data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytest4, yPredtest4), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Testing: Accuracy Score: {0}'.format(skm.accuracy_score(ytest4, yPredtest4)) 
plt.title(all_sample_title, size = 15);
```
