```python
import pandas as pd
from sklearn.model_selection import KFold

# Load your data into dataKFold
dataKFold = df

# Initialize KFold
kf = KFold(n_splits=4, shuffle=True, random_state=42)  # 4 splits for k-fold

# Create a counter for naming the files
fold_counter = 1

# Loop through the splits
for train_index, test_index in kf.split(dataKFold):
    # Split the data into training and testing sets
    train_data = dataKFold.iloc[train_index]
    test_data = dataKFold.iloc[test_index]
    
    # Save the training and testing data for this fold to separate files
    train_data.to_csv(f'C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Train_fold_{fold_counter}.csv', index=False)
    test_data.to_csv(f'C:/Users/User/OneDrive/Documents/Group Project/group_project_dataIT-Test_fold_{fold_counter}.csv', index=False)
    
    # Increment the fold counter
    fold_counter += 1
```

```python
#Train-1
# Convert dataframe to array 
xtrain1 = xtrain1.values
ytrain1 = ytrain1.values 
ytrain1 = ytrain1.ravel() 

# K-Nearest Neighbor
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
from sklearn.neighbors import KNeighborsClassifier 
ModelAKU = KNeighborsClassifier(n_neighbors=67, metric='hamming', weights='distance', algorithm='ball_tree', leaf_size=20) 

ModelAKU.fit(xtrain1, ytrain1) 
yPredtrain1 = ModelAKU.predict(xtrain1) 

# Testing & performance analysis -Training data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytrain1, yPredtrain1), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Training: Accuracy Score: {0}'.format(skm.accuracy_score(ytrain1, yPredtrain1)) 
plt.title(all_sample_title, size = 15);
```

```python
#Test-1
# Convert dataframe to array 
xtest1 = xtest1.values 
ytest1 = ytest1.values 
ytest1 = ytest1.ravel() 

yPredtest1 = ModelAKU.predict(xtest1) 

# Testing & performance analysis -Testing data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(ytest1, yPredtest1), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Testing: Accuracy Score: {0}'.format(skm.accuracy_score(ytest1, yPredtest1)) 
plt.title(all_sample_title, size = 15);
```
